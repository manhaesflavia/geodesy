Dados 
--------- 

coordenada no primeiro sistema (lat1,long1)


*   a1 - semi-eixo maior no sistema 1
*   f1 - achatamento no sistema 1
*   a2 - semi-eixo maior no sistema 2
*   f2 - achametamento no sistema 2


----

Parâmetros de transformação entre s1 e s2
*   delX12
*   delY12
*   delZ12


```python
# Função para transformar a coordenada GMS para GD

def dms2dd(degrees, minutes, seconds, direction):
    dd = float(degrees) + float(minutes)/60 + float(seconds)/(60*60);
    if direction == 'W' or direction == 'S':
        dd *= -1
    return dd

def dd2dms(deg):
    d = int(deg)
    md = abs(deg - d) * 60
    m = int(md)
    sd = (md - m) * 60
    return [d, m, sd]

import numpy as np
import math

```

```python
dms2dd(49, 13, 51.4116, "W")
```

```python


def edMolodenskii (a1,f1,a2,f2,delX12,delY12, delZ12,latGD,longGD):

  # Primeira Excentricidade ao quadrado no S1 (e2_1)
  e2_1=2*f1-f1**2
  print(round(e2_1,10),"Primeira Excentricidade ao quadrado no S1 (e2_1)")

  # Segunda Excentricidade ao quadrado no S1 (e2´_1)
  ee2_1= e2_1/(1-e2_1)
  print(round(ee2_1,10),"Segunda Excentricidade ao quadrado no S1 (e2´_1)")

  # Grande Normal no S1
  N1=(a1/math.sqrt(1-e2_1*((np.sin(np.radians(latGD))**2))))
  print(round(N1,3),"Grande Normal no S1")

  # Raio de Curvatura na seção meridiana no S1 (M1)
  numerador = a1*(1-e2_1)
  denoP1=(1-e2_1*((np.sin(np.radians(latGD))**2)))
  M1=(numerador/(denoP1**(3/2)))
  print(round(M1,3),"Raio de Curvatura na seção meridiana no S1")

  # Variação entre os semi-eixo maior (delA=a2-a1)
  delA=a2-a1
  print(delA,"Variação entre os semi-eixo maior (delA=a2-a1)")

  # Variação entre os achatamentos (delF=f2-f1)
  delF=f2-f1
  print(delF,"Variação entre os achatamentos (delF=f2-f1)")

  # Cálculo do Termo 1
  termo1=(((a1*delF)+(f1*delA))*(np.sin(np.radians(2*latGD))))
  print(round(termo1,10),"termo1")

  # Cálculo do Termo 2
  termo2=delX12*np.sin(np.radians(latGD))*np.cos(np.radians(longGD))
  print(round(termo2,8),"termo2")

  # Cálculo do Termo 3
  termo3 = delY12*np.sin(np.radians(latGD))*np.sin(np.radians(longGD))
  print(round(termo3,8),"termo3")

  # Cálculo do Termo 4
  termo4=delZ12*np.cos(np.radians(latGD))
  print(round(termo4,8),"termo4")

  # Cálculo da variação da latitude (delLat)
  delLat=((1/M1)*(termo1-termo2-termo3+termo4))
  print(delLat,"variação da latitude (delLat)")

  # Cálculo do Termo 5
  termo5=(1/(N1*np.cos(np.radians(latGD))))
  print(round(termo5,8),"termo5")

  # Cálculo do Termo 6
  termo6=delX12*np.sin(np.radians(longGD))*(-1)
  print(round(termo6,8),"termo6")

  #Cálculo do Termo 7
  termo7=delY12*np.cos(np.radians(longGD))
  print(round(termo7,9),"termo7") 

  # Cálculo da variação da longitude (delLong)
  delLong=termo5*(termo6+termo7)
  print(delLong,"variação da longitude (delLong)")

  #Cálculo do Termo 8
  termo8=(((a1*delF)+(f1*delA))*(np.sin(np.radians(latGD))**2))
  print(round(termo8,10),"termo8")

  # Cálculo do Termo 9
  termo9=delX12*np.cos(np.radians(latGD))*np.cos(np.radians(longGD))
  print(round(termo9,9),"termo9")

  # Cálculo do Termo 10
  termo10=delY12*np.cos(np.radians(latGD))*np.sin(np.radians(longGD))
  print(round(termo10,9),"termo10")

  # Cálculo do Termo 11
  termo11=delZ12*np.sin(np.radians(latGD))
  print(round(termo11,9),"termo11")

  # Diferença de Geondulação (delN)
  delN=termo8-delA+termo9+termo10+termo11
  print(round(delN,9),"delN")

  # Coordenada no sistema 2

  latGD2=(latGD+(delLat*(180/math.pi)))
  longGD2=(longGD+(delLong*(180/math.pi)))
  return [dd2dms(latGD2),dd2dms(longGD2)]

```
